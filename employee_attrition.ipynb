```python
import pandas as pd
```

```python
df_emp=pd.read_csv('/content/Employee-Attrition - Employee-Attrition.csv')
```

```python
df_emp.head()
```

```python
df_emp.dtypes
```

```python
df_emp.isna().sum()
```

```python
df_emp['Attrition']=df_emp['Attrition'].replace({'Yes':1,'No':0})
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoders={}
for col in df_emp.select_dtypes(include=['object']).columns:
  le=LabelEncoder()
  df_emp[col]=le.fit_transform(df_emp[col])
  label_encoders[col]=le
```

```python
from sklearn.preprocessing import StandardScaler
columns_to_scale=['DailyRate', 'HourlyRate', 'MonthlyIncome', 'MonthlyRate']
scaler=StandardScaler()
df_emp[columns_to_scale] = scaler.fit_transform(df_emp[columns_to_scale])
```

```python
df_emp.head()
```

```python
df_emp.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,20))
sns.heatmap(df_emp.corr(),annot=True,cmap="coolwarm",fmt="2f",linewidths=2)
plt.title("Correlation Matrix",fontsize=20)
plt.show()
```

```python
from sklearn.feature_selection import mutual_info_classif
x=df_emp.drop(columns=["Attrition"])
y=df_emp["Attrition"]
mi_scores = mutual_info_classif(x, y, discrete_features='auto')
mi_scores_series = pd.Series(mi_scores, index=x.columns).sort_values(ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x=mi_scores_series, y=mi_scores_series.index, palette="viridis")
plt.xlabel("Mutual Information Score")
plt.ylabel("Feature")
plt.title("Feature Importance based on Mutual Information")
plt.show()
```

```python
top_features = ['MonthlyIncome', 'YearsAtCompany', 'JobRole', 'YearsInCurrentRole', 'Age', 'JobLevel', 'TotalWorkingYears', 'StockOptionLevel', 'OverTime', 'Education']
X = df_emp[top_features]
y = df_emp['Attrition']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
lr=LogisticRegression()
algo1=lr.fit(x_train,y_train)
```

```python
print("Accuracy Score:",accuracy_score(y_test,algo1.predict(x_test)))
print("Classification Report:\n",classification_report(y_test,algo1.predict(x_test)))
print("Confusion Matrix:\n",confusion_matrix(y_test,algo1.predict(x_test)))
```
